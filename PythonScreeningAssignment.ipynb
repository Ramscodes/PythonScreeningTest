2.Demonstrate use of abstract class, multiple inheritance and decorator in
python using examples.

2.1  Abstract class

```python
from abc import ABC, abstractmethod
```

```python
class Abstract:
    @abstractmethod
    def show(self):
        pass
```

```python
class Regular1(Abstract):
    def show(self):
        print("Inside Regular1")
```

```python
class Regular2(Abstract):
    def show(self):
        print("Inside Regular2")
```

```python
r1 = Regular1()
r1.show()
```

```python
r2 = Regular2()
r2.show()
```

2.2 Multiple Inheritance

```python
class A:
    def __init__(self,a,b):
        self.a =a
        self.b=b
    def showA(self):
        print("inside A class",self.a,self.b)
class B():
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def showB(self):
        print("inside B class",self.x,self.y)
class C(A,B):
    pass
class D(B,A):
    pass
```

```python
c = C(1,2)#a,b from class A
d = D(3,4)#x,y from class B
```

```python
c.showA()
d.showB()
```
